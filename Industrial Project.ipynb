## Checkpoint 6- Plots

#### 1.  A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type. This chart should make it clear that a certain funding type (FT) is best suited for Spark Funds

```python
df1=master_frame2[(master_frame2['funding_round_type']=='venture')|(master_frame2['funding_round_type']=='seed')|(master_frame2['funding_round_type']=='private_equity')]

df2=df1.groupby(by='funding_round_type')['raised_amount_usd'].sum().sort_values(ascending=False).to_frame().reset_index()  
```

```python
df3=df1.groupby(by='funding_round_type')['raised_amount_usd'].mean().sort_values(ascending=False).to_frame().reset_index()  
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.yscale('log')
sb.barplot(df2['funding_round_type'],df2['raised_amount_usd'],palette='ocean')

plt.subplot(122)
plt.yscale('log')
sb.barplot(df3['funding_round_type'],df3['raised_amount_usd'],palette='ocean')
plt.show()
```
